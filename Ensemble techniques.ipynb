```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv(r"E:\My Work\Top_Dataset\Brain Stroke.csv")
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
en=LabelEncoder()
```

```python
dataset["gender"]=en.fit_transform(dataset["gender"])
dataset["ever_married"]=en.fit_transform(dataset["ever_married"])
dataset["work_type"]=en.fit_transform(dataset["work_type"])
dataset["Residence_type"]=en.fit_transform(dataset["Residence_type"])
dataset["smoking_status"]=en.fit_transform(dataset["smoking_status"])
```

```python
x=dataset.iloc[:,:-1].values # independent 
y=dataset.iloc[:,-1].values  # dependent 
```

```python
from imblearn.over_sampling import SMOTE
sms=SMOTE()
x_data,y_data=sms.fit_resample(x,y)
```

```python
from sklearn.preprocessing import RobustScaler
Rb=RobustScaler()
x_robust_scaler=Rb.fit_transform(x_data)
```

```python
x_robust_scaler
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_robust_scaler,y_data,test_size=0.2,random_state=11)
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score

log=LogisticRegression()
log.fit(x_train,y_train)

y_log=log.predict(x_test)
ac_log=accuracy_score(y_log,y_test)*100
ac_log
```

```python
y_log
```

```python
from sklearn.svm import SVC 
from sklearn.metrics import accuracy_score

sv=SVC()
sv.fit(x_train,y_train)

y_sv=sv.predict(x_test)
ac_sv=accuracy_score(y_sv,y_test)*100
ac_sv
```

```python
y_sv
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn=KNeighborsClassifier()
knn.fit(x_train,y_train)

y_knn=knn.predict(x_test)
ac_knn=accuracy_score(y_knn,y_test)*100
ac_knn
```

```python
y_knn
```

```python
# Averging
y_all=(ac_log+ac_sv+ac_knn)/3
```

```python
y_all
```

```python
y_all_pred=(y_log+y_sv+y_knn)/3
y_all_pred
```

```python
# Maxvoting Classifier

from sklearn.ensemble import VotingClassifier
max_vote=VotingClassifier(estimators=[('Logistic Regression', log), ('SVM', sv), ('KNN', knn)])
max_vote.fit(x_train,y_train)
max_vote_pred=max_vote.predict(x_test)
max_vote_pred
```

```python
ac_max_vote=accuracy_score(max_vote_pred,y_test)*100
ac_max_vote
```

```python
# Bagging Classifier
from sklearn.ensemble import BaggingClassifier
bg=BaggingClassifier(estimator=knn,n_estimators=100, random_state=99)
bg.fit(x_train,y_train)

bg_pred=bg.predict(x_test)
bg_pred
```

```python
ac_bg=accuracy_score(bg_pred,y_test)*100
ac_bg
```

```python
# RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
rms=RandomForestClassifier(n_estimators=100)
rms.fit(x_train,y_train)
y_rms=rms.predict(x_test)
y_rms
```

```python
# [1, 0, 1, ..., 1, 0, 0], ---rms
ac_rms=accuracy_score(y_rms,y_test)*100
ac_rms
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gboost=GradientBoostingClassifier(loss='log_loss', learning_rate=0.1, n_estimators=100)
gboost.fit(x_train,y_train)
y_boost=gboost.predict(x_test)
ac_boost=accuracy_score(y_boost,y_test)*100
ac_boost
```
