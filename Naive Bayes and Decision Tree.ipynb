```python
import pandas as pd
import numpy as np
dataset=pd.read_csv(r"E:\My Work\Top_Dataset\Brain Stroke.csv")
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
en=LabelEncoder()
dataset["gender"]=en.fit_transform(dataset["gender"])
dataset["ever_married"]=en.fit_transform(dataset["ever_married"])
dataset["work_type"]=en.fit_transform(dataset["work_type"])
dataset["Residence_type"]=en.fit_transform(dataset["Residence_type"])
dataset["smoking_status"]=en.fit_transform(dataset["smoking_status"])
```

```python
x=dataset.iloc[:,:-1].values 
y=dataset.iloc[:,-1].values  
```

```python
from imblearn.over_sampling import SMOTE
sms=SMOTE()
x_data,y_data=sms.fit_resample(x,y)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
x_scale=mms.fit_transform(x_data)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scale,y_data,test_size=0.2,random_state=211)
```

```python
from sklearn.metrics import accuracy_score 
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
nb_pred=nb.predict(x_test)
ac_nb=accuracy_score(nb_pred,y_test)*100
ac_nb
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='gini',splitter='best',max_depth=2)
dt.fit(x_train,y_train)
dt_pred=dt.predict(x_test)
ac_dt=accuracy_score(dt_pred,y_test)*100
ac_dt
```
