```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold

from sklearn.linear_model import LogisticRegression 

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict

from sklearn.metrics import accuracy_score
from collections import Counter
```

### Create X and Y Values

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

### Work on the Data Imbalancing Technique

```python
rm=RandomOverSampler()
x_data,y_data=rm.fit_resample(x,y)
```

### Define Stratify Fold Size with K FOLD Cross Validation Technique

```python
ksf=StratifiedKFold(n_splits=5,random_state=3, shuffle=True)
ksf.get_n_splits(x_data,y_data)
print(ksf)
```

```python
from sklearn.linear_model import LogisticRegression
l1=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
for train_data,test_data in ksf.split(x_data,y_data):
    x_train,x_test=x_data[train_data],x_data[test_data]
    y_train,y_test=y_data[train_data],y_data[test_data]

    score=cross_val_score(l1,x_train,y_train,cv=kf)
    y_pred=cross_val_predict(l1,x_test,y_test,cv=kf)
    
    print("Train : ",y_train.shape," Test: ",y_test.shape)
    print("Train : ",Counter(y_train)," Test: ",Counter(y_test))
    print(score)
    print(y_pred)
    print(np.mean(score)*100)
```

## This is Stratify K Fold Cross Validation

```python
for train_data,test_data in ksf.split(x,y):
    x_train,x_test=x_data[train_data],x_data[test_data]
    y_train,y_test=y_data[train_data],y_data[test_data]
    
    score=cross_val_score(l1,x_train,y_train,cv=ksf)
    y_pred=cross_val_predict(l1,x_test,y_test,cv=ksf)
    
    print("-------My Fold Accuracy Score by Cross Validation Technique-------")
    print(score)
    
    print("**************************************************************")
    print(y_pred)
    print("------------------------------------------------------")
    print("Check Dataset is Balanced")
    import numpy as np
    (unique, counts) = np.unique(y_test, return_counts=True)
    frequencies = np.asarray((unique, counts)).T
    print(frequencies)
    print("--------------------------------------------------------")
    
    from sklearn.metrics import confusion_matrix,accuracy_score
    ac=accuracy_score(y_pred,y_test)*100
    print("Accuracy Score: ",ac)

    print("---------------------------------------------------------")

    cm=confusion_matrix(y_pred,y_test)
    print("This is Confusion Metrics: ")
    print(cm)

    
```
